```python
# Import dependances.
import json
import pandas as pd
import numpy as np
```

```python
# Load CDC dataframe.
df_cdc = pd.read_json('data/cdc_state_5_years.json')
```

```python
df_cdc.head()
```

```python
# Sort by columns year, week, and state.
df_cdc = df_cdc.sort_values(by=['year', 'state', 'week'])
```

```python
df_cdc.head()
```

```python
# Load HHS dataframe.
df_hhs = pd.read_json('data/hhs_state_5_years.json')
```

```python
# Reduce columns to those that are needed.
df_hhs = df_hhs.filter(['count','name','percentage', 'week'], axis=1)
# df_hhs.head()
```

```python
df_hhs.head()
```

```python
# Merge with outer join the HHS and CDC dataframes.
df1 = pd.merge(df_cdc, df_hhs, how='outer', on=['year', 'state', 'week'])
df1 = df1.sort_values(by=['year', 'state', 'week'])
df1.fillna(0, inplace=True)
# result
```

```python
df1.head()
```

```python
# Write the merged datafram to the json file.
df1.to_json('data/plot_state_5_years.json')
```
